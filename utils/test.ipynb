```python
import os

# base path: one folder up from notebook, then into data/SynthText/raw/...
synth_root = os.path.abspath(
    os.path.join("..", "data", "SynthText", "raw", "wendlerc___captioned_synth_text",
                 "default", "0.0.0", "62a4664ee9b58ef472d85fae6e6abbad9ec16c14")
)

print(os.path.exists(synth_root))  # should be True

# list arrow files
arrow_files = [
    os.path.join(synth_root, f)
    for f in os.listdir(synth_root)
    if f.endswith(".arrow")
]

print(f"Found {len(arrow_files)} Arrow files")
print(arrow_files[:5])
```

```python
from datasets import load_dataset

# load the first file
ds = load_dataset("arrow", data_files=arrow_files[0], split="train")

# inspect columns
print(ds.column_names)
print(ds[0])
```
